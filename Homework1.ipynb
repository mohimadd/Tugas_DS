```python
# nomor 1 
# buat kolom fullname v
# dan disisipkan setelah kolom last name
# isinya gabungan First Name - Last name pakai title case

df_sheet1 ['Full Name'] = df_sheet1 ['First Name'] + " " + df_sheet1['Last Name']
df_sheet1
```

```python
# nomor 1 
# buat kolom fullname v
# dan disisipkan setelah kolom last name v
# isinya gabungan First Name - Last name pakai title case

df_sheet1.columns
df_sheet1 = df_sheet1.loc[:,['Paid Date', 'Order Number', 'First Name', 'Last Name','Full Name','Meta Category',
       'Product Name', 'Transaction Amount', 'Seller Discount',
       'Sales Discount', 'Delivery Fee', 'Other Discount']]
df_sheet1.head()
```

```python
# nomor 1 
# buat kolom fullname v
# dan disisipkan setelah kolom last name v
# isinya gabungan First Name - Last name pakai title case v

df_sheet1['Full Name'] = df_sheet1['Full Name'].str.upper().str.title()
```

```python
df_sheet1.head()
```

```python
# nomor 2
# buat kolom "GMV" v
# letakan di sebelah kolom "Seller Discount"
# isinya adalah "Transaction-Amount - "Seller Discount" + "Delivery Fee" v

df_sheet1 ["GMV"] = df_sheet1 ['Transaction Amount'] - df_sheet1['Seller Discount'] + df_sheet1["Delivery Fee"]
df_sheet1
```

```python
# nomor 2
# buat kolom "GMV" v
# letakan di sebelah kolom "Seller Discount" v
# isinya adalah "Transaction-Amount - "Seller Discount" + "Delivery Fee" v

df_sheet1.columns
df_sheet1 = df_sheet1.loc[:, ['Paid Date', 'Order Number', 'First Name', 'Last Name', 'Full Name',
       'Meta Category', 'Product Name', 'Transaction Amount',
       'Seller Discount', 'Sales Discount','GMV','Delivery Fee', 'Other Discount']]
df_sheet1.head()
```

```python
df_sheet1['Meta Category'].nunique
df_sheet1['Meta Category'].unique()
```

```python
# nomor 3
# kelompokkan meta category menjadi group 1 n 2 -> buat kolom tambahan yang berisikan keterangan group 1 atau 2 v
# buat df yang membuat pivot table dari data frame paid transaction untuk menampilkan total nilai gmv perbulan untuk setiap group 

def grup_kategori (MetaCategory):
    group_1 = ['Home/ Food','Sports/ Hobi/ Otomotif','Fashion', 'Beauty/ Health']
    group_2 = ['Gadget/ Komputer', 'Elektronik', 'Service/ Mokado', 'Babies/ Kids']

    if MetaCategory in group_1:
        return "1"
    elif MetaCategory in group_2:
        return "2"
```

```python
df_sheet1['Group'] = df_sheet1['Meta Category'].apply(grup_kategori)
```

```python
df_sheet1.head()
```

```python
# nomor 3
# kelompokkan meta category menjadi group 1 n 2 -> buat kolom tambahan yang berisikan keterangan group 1 atau 2 v
# buat df yang membuat pivot table dari data frame paid transaction untuk menampilkan total nilai gmv perbulan untuk setiap group 

# ubah Paid date ke Date Time
df_sheet1['Paid Date'] = pd.to_datetime(df_sheet1['Paid Date'], format = '%Y%m%d')
```

```python
df_sheet1.head()
```

```python
# ekstraksi data paid date ke bulan

df_sheet1['Paid Date M'] = df_sheet1['Paid Date'].dt.month
df_sheet1.head()
```

```python
# ganti bulan dari angka ke nama

df_sheet1['Paid Date M'] = df_sheet1['Paid Date M'].map({1:'Januari', 2 : 'Februari', 3 : 'Maret', 4 : 'April', 5 : 'Mei', 6 : 'Juni', 7 : 'Juli', 8 : 'Agustus', 9 : 'September', 10 : 'Oktober', 11 : 'November', 12 : 'Desember'})
```

```python
# nomor 3
# kelompokkan meta category menjadi group 1 n 2 -> buat kolom tambahan yang berisikan keterangan group 1 atau 2 v
# buat df yang membuat pivot table dari data frame paid transaction untuk menampilkan total nilai gmv perbulan untuk setiap group 

df_sheet1.head()
```

```python
# buat pivot table 

df_pivot = pd.pivot_table(df_sheet1, index = ['Group', 'Paid Date M'], values = 'GMV', aggfunc = 'sum')
df_pivot
```

```python
# nomor 4 
# Buatlah python statement untuk mencari seller mana yang memiliki total GMV paling tinggi di bulan agustus 2017
# gunakan sheet (SELLER)


df_join = pd.merge(df_sheet1,df_sheet2, on ='Order Number', how = 'inner')
df_join
```

```python
df_join.info()
```

```python
# filer hanya bulan agustus

df_agus = df_join[df_join['Paid Date M'] == 'Agustus']
df_agus.head()
```

```python
#Group by
df_agus.groupby(['Seller']).agg({'GMV' : 'sum'}).sort_values('GMV', ascending = False)
```

```python
# nomor 5
# mencari seller yang paling banyak melakukan transaksi di bulan september untuk kategori "Fashion"

# create filter - september & fashion
df_sept = df_join[(df_join['Paid Date M'] == 'September') & (df_join['Meta Category'] == "Fashion")]
df_sept.head()
```

```python
df_sept.groupby('Seller').agg({'Transaction Amount' : 'sum'}).sort_values('Transaction Amount', ascending = False)
```
